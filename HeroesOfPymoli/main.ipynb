```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

```python
#Display the total number of players
player_count = purchase_data['SN'].value_counts().count()
column = ['Total Players']
pd.DataFrame([player_count],columns=column)
```

```python
#Here we will review a total purchasing analysis
#Number of unique items
unique_purchases = len(purchase_data['Item ID'].unique())

#Average purchase price
average_pur_price = purchase_data['Price'].mean()

#Total number of purchases
total_purchases = purchase_data['Purchase ID'].count()

#Total revenue
total_revenue = purchase_data['Price'].sum()

total_pur_analysis = {'Number of Unique Items':[unique_purchases],'Average Price':[average_pur_price],
     'Number of Purchases':total_purchases,'Total Revenue':total_revenue}

pd.DataFrame(total_pur_analysis)
```

```python
#Gender Demographics
gen_dem_df = purchase_data[['SN','Gender']].drop_duplicates()
player_count
#Percentage and Count of Male Players
male_count = gen_dem_df.loc[gen_dem_df['Gender']=='Male', :]
male_count = len(male_count)
male_percent = male_count/player_count*100
male_percent = male_percent.round(2)

#Percentage and Count of Female Players
female_count = gen_dem_df.loc[gen_dem_df['Gender']=='Female', :]
female_count = len(female_count)
female_percent = female_count/player_count*100
female_percent = female_percent.round(2)


#Percentage and Count of Other / Non-Disclosed
other_count = gen_dem_df.loc[gen_dem_df['Gender']=='Other / Non-Disclosed']
other_count = len(other_count)
other_percent = other_count/player_count*100
other_percent = other_percent.round(2)

df = {'Total Count':[male_count,female_count,other_count],
      'Percentage of Players':[male_percent,female_percent,other_percent]}
rows = ['Male','Female','Other / Non-Disclosed']
df = pd.DataFrame(df)
df
```

```python
#Purchasing Analysis (Gender)
#Purchase Count, Average Purchase Price, Total Purchase Value, Average Purchase Total per Person by Gender
gender_info = purchase_data.groupby('Gender')
total_gen_count = gender_info.nunique()["SN"]

pur_count = gender_info['Purchase ID'].count()

avg_purchase = gender_info['Price'].mean()

total_value = gender_info['Price'].sum()

avg_pur_person = total_value/total_gen_count

#Create a summary data frame to hold the results
gender_df = pd.DataFrame({'Purchase Count':pur_count,'Average Purchase Price':avg_purchase,
             'Total Purchase Value':total_value,'Avg Total Purchase per Person':avg_pur_person})
#Display the summary data frame
gender_df
```

```python
# Age Demographics
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Demographics"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)

age_group = purchase_data.groupby("Age Demographics")
total_age_count = age_group.nunique()["SN"]
#Purchase Count
age_pur_count = age_group['Purchase ID'].count()
#Average Purchase Price
age_pur_avg = age_group['Price'].mean()
#Total Purchase Value
age_pur_total = age_group['Price'].sum()
#Average Purchase Total per Person by Age Group
avg_per_person = age_pur_total/total_age_count

age_dem_df = pd.DataFrame({'Purchase Count': age_pur_count, 'Average Purchae Price': age_pur_avg,
                          'Total Purchase Value': age_pur_total,'Avg Total Purchase per Person': avg_per_person})
age_dem_df
```

```python
#Top Spenders
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#SN
top_spenders = purchase_data.groupby('SN')

#Purchase Count
spender_count = top_spenders["Purchase ID"].count()
#Average Purchase Price
spender_avg = top_spenders["Price"].mean()
#Total Purchase Value
spender_total_value = top_spenders["Price"].sum()

spenders_df = pd.DataFrame({'Purchase Count':spender_count,'Average Purchase Price':spender_avg,
                            'Total Purchase Value':spender_total_value})

spenders_df = spenders_df.sort_values(["Total Purchase Value"], ascending=False)
spenders_df.head(5)
```

```python
#Most Popular Items
#Identify the 5 most popular items by purchase count, then list (in a table):
#Retrieve the Item ID, Item Name, and Item Price columns
items = purchase_data[['Item ID','Item Name','Price']]
#Group by Item ID and Item Name. 
popular_items = items.groupby(['Item ID','Item Name'])

#Perform calculations to obtain purchase count, average item price, and total purchase value

popular_pur_count = popular_items['Price'].count()

popular_total_value = popular_items['Price'].sum()

popular_item_price = popular_total_value/popular_pur_count

#Create a summary data frame to hold the results
popular_df = pd.DataFrame({'Purchase Count':popular_pur_count,'Item Price':popular_item_price,
                           'Total Purchase Value':popular_total_value})

popular_df = popular_df.sort_values(["Purchase Count"], ascending=False)

popular_df.head(5)
```

```python
#Most Profitable Items
#Identify the 5 most profitable items by total purchase value, then list (in a table):
#Sort the above table by total purchase value in descending order
profitable_df = popular_df.sort_values(["Total Purchase Value"], ascending=False)
profitable_df.head(5)
```
