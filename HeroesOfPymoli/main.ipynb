```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(20)
```

```python
#Display the total number of players
player_count = purchase_data['SN'].value_counts().count()
column = ['Total Players']
pd.DataFrame([player_count],columns=column)
```

```python
#Here we will review a total purchasing analysis
#Number of unique items
unique_purchases = len(purchase_data['Price'].unique())

#Average purchase price
average_pur_price = purchase_data['Price'].mean()

#Total number of purchases
total_purchases = purchase_data['Price'].count()

#Total revenue
total_revenue = purchase_data['Price'].sum()

total_pur_analysis = {'Number of Unique Items':[unique_purchases],'Average Price':[average_pur_price],
     'Number of Purchases':total_purchases,'Total Revenue':total_revenue}

pd.DataFrame(total_pur_analysis)
```

```python
#Gender Demographics
gen_dem_df = purchase_data[['SN','Gender']].drop_duplicates()
player_count
#Percentage and Count of Male Players
male_count = gen_dem_df.loc[gen_dem_df['Gender']=='Male', :]
male_count = len(male_count)
male_percent = male_count/player_count*100
male_percent = male_percent.round(2)

#Percentage and Count of Female Players
female_count = gen_dem_df.loc[gen_dem_df['Gender']=='Female', :]
female_count = len(female_count)
female_percent = female_count/player_count*100
female_percent = female_percent.round(2)


#Percentage and Count of Other / Non-Disclosed
other_count = gen_dem_df.loc[gen_dem_df['Gender']=='Other / Non-Disclosed']
other_count = len(other_count)
other_percent = other_count/player_count*100
other_percent = other_percent.round(2)

df = {'Total Count':[male_count,female_count,other_count],
      'Percentage of Players':[male_percent,female_percent,other_percent]}
rows = ['Male','Female','Other / Non-Disclosed']
df = pd.DataFrame(df)
df
```

```python
#Purchasing Analysis (Gender)
#Purchase Count, Average Purchase Price, Total Purchase Value, Average Purchase Total per Person by Gender
female_count = purchase_data.loc[purchase_data['Gender']=='Female', :]
pur_count_fem = female_count['Price'].count()
avg_fem_purchase = female_count['Price'].mean()
total_value_fem = female_count['Price'].sum()
#avg_total_per_fem = female_count.groupby(['SN']).mean()

male_count = purchase_data.loc[purchase_data['Gender']=='Male', :]
pur_count_male = male_count['Price'].count()
avg_male_purchase = male_count['Price'].mean()
total_value_male = male_count['Price'].sum()
#avg_total_per_fem = female_count.groupby(['SN']).mean()

other_count = purchase_data.loc[purchase_data['Gender']=='Other / Non-Disclosed', :]
pur_count_other = other_count['Price'].count()
avg_other_purchase = other_count['Price'].mean()
total_value_other = other_count['Price'].sum()
#avg_total_per_fem = female_count.groupby(['SN']).mean()

#Create a summary data frame to hold the results
gender_df = {'Purchase Count':[pur_count_fem,pur_count_male,pur_count_other],
            'Average Purchase Price':[avg_fem_purchase,avg_male_purchase,avg_other_purchase],
            'Total Purchase Value':[total_value_fem,total_value_male,total_value_other]}
gender_df = pd.DataFrame(gender_df)
#Display the summary data frame
gender_df
```
